### Change Columns

```python
new_columns = {
    'Survived' : 'SURVIVED',
    'Name' : 'NAME',
    'Pclass' : 'PCLASS',
    'Sex' : 'GENDER',
    'Age' : 'AGE',
    'Siblings/Spouses Aboard' : 'SIBSA',
    'Parents/Children Aboard' : 'PARCA',
    'Fare' : 'PRICE'
}
```

```python
df = df.rename(columns = new_columns)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
new = df["NAME"].str.split(" ", n = 2, expand = True)
df['Title'] = new[0]
df["First Name"]= new[1]
df["Last Name"]= new[2]
```

### Concatenate

```python
df1 = df[['SURVIVED', 'PCLASS', 'NAME']]
df2 = df[['GENDER', 'AGE', 'SIBSA']]
df3 = df[['PARCA', 'PRICE']]

list_df = [df1, df2, df3]
```
