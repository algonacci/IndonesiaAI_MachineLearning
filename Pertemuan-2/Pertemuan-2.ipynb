### Data Visualization

#### Histogram for Distribution

```python
df['sepal_length'].hist() 
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.show()
```

```python
df['petal_length'].hist() 
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.show()
```

#### Using Klib

```python
klib.dist_plot(df)
plt.show()
```

```python
df['species'].hist()
plt.show()
```

#### Using Seaborn

```python
sns.countplot(x=df['species'])
plt.show()
```

### Scatter plot - Distribution & Relation

```python
df.plot(kind='scatter',x='sepal_length',y='sepal_width', c='lightblue')
plt.show()
```

```python
sns.scatterplot(data=df, x='sepal_length',y='sepal_width', hue='species')
plt.show()
```

```python
df.plot(kind='scatter',x='petal_length',y='petal_width', c='lightblue')
plt.show()
```

```python
sns.scatterplot(data= df, x='petal_length',y='petal_width', hue='species', palette='dark')
plt.show()
```

### Displot - Distribution

```python
sns.displot(df['sepal_length'],kde=True)
plt.show()
```

```python
sns.kdeplot(x=df['sepal_length'], hue=df['species'], shade=True)
```

```python
sns.displot(x=df['sepal_length'], kde=True, bins=20)
plt.show()
```

### Boxplot - Distribution and Comparison

```python
sns.boxplot(data=df)
plt.xlabel('Features')
plt.ylabel('Length (cm)')
plt.show()
```

```python
sns.boxplot(data=df, y='petal_length',x='species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

### Pairplot - Relation, Distribution, Comparison

```python
sns.pairplot(data=df)
plt.show()
```

```python
sns.pairplot(data=df, hue='species')
plt.show()
```

### Heatmap - Correlation

```python
df.corr()
```

```python
sns.heatmap(data=df.corr())
plt.show()
```

#### With `annot=True`

```python
sns.heatmap(data=df.corr(), annot=True)
plt.show()
```
